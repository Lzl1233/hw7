# Q1: Derivations

## 1. Posterior Distribution of $\beta$

Given the linear regression model:

\[
p(\beta | y, \mathbf{X}, \Sigma = \sigma^2 I) \propto \exp \left( -\frac{1}{2} (y - \mathbf{X} \beta)^T \Sigma^{-1} (y - \mathbf{X} \beta) \right)
\]

and a prior:

\[
\beta \sim \mathcal{MVN}(\beta_0, \Sigma_{\beta})
\]

We derive the posterior distribution of $\beta$ as follows.

\[
p(\beta | y, \mathbf{X}, \sigma^2) \sim \mathcal{MVN}(\mu_{\beta}, \Sigma_{\beta})
\]

where

\[
\Sigma_{\beta}^{-1} = \Sigma_{\beta_0}^{-1} + \frac{1}{\sigma^2} \mathbf{X}^T \mathbf{X}
\]

\[
\mu_{\beta} = \Sigma_{\beta} \left( \Sigma_{\beta_0}^{-1} \beta_0 + \frac{1}{\sigma^2} \mathbf{X}^T y \right)
\]

## 2. Posterior Distribution of $\sigma^2$

Assuming an inverse-gamma prior:

\[
\sigma^2 \sim \text{Inv-Gamma}(\alpha^*, \beta^*)
\]

The posterior follows:

\[
p(\sigma^2 | y, \mathbf{X}) \sim \text{Inv-Gamma}(\alpha^* + \frac{n}{2}, \beta^* + \frac{1}{2} (y - \mathbf{X} \beta)^T (y - \mathbf{X} \beta))
\]

## [Optional] Q1 Extra: Posterior for Normal and Laplace Priors

For a normal prior on $\beta$:

\[
\beta \sim \mathcal{N}(0, \lambda I)
\]

The log-posterior:

\[
\log p(\beta | y, X) \propto -\frac{1}{2} \| y - X \beta \|^2 - \frac{\lambda}{2} \|\beta\|^2
\]

which resembles **ridge regression**.

For a Laplace prior:

\[
p(\beta) \propto \exp(-\lambda |\beta|)
\]

The log-posterior:

\[
\log p(\beta | y, X) \propto -\frac{1}{2} \| y - X \beta \|^2 - \lambda \|\beta\|_1
\]

which resembles **lasso regression**.


```python
import numpy as np
import scipy.stats as stats

def posterior_beta(X, y, beta_0, Sigma_beta, sigma2):
    """
    Compute posterior parameters for beta in Bayesian linear regression
    """
    Sigma_beta_inv = np.linalg.inv(Sigma_beta)
    Sigma_post_inv = Sigma_beta_inv + (1/sigma2) * X.T @ X
    Sigma_post = np.linalg.inv(Sigma_post_inv)
    mu_post = Sigma_post @ (Sigma_beta_inv @ beta_0 + (1/sigma2) * X.T @ y)
    return mu_post, Sigma_post

def posterior_sigma2(y, X, beta, alpha_star, beta_star):
    """
    Compute posterior parameters for sigma^2 given inverse-gamma prior
    """
    n = len(y)
    residuals = y - X @ beta
    alpha_post = alpha_star + n / 2
    beta_post = beta_star + 0.5 * (residuals.T @ residuals)
    return alpha_post, beta_post

# Example usage
np.random.seed(42)
n, p = 100, 5
X = np.random.randn(n, p)
beta_true = np.random.randn(p)
sigma2_true = 2.0
y = X @ beta_true + np.random.randn(n) * np.sqrt(sigma2_true)

# Prior settings
beta_0 = np.zeros(p)
Sigma_beta = np.eye(p)
alpha_star, beta_star = 2, 2

# Compute posterior parameters
mu_beta_post, Sigma_beta_post = posterior_beta(X, y, beta_0, Sigma_beta, sigma2_true)
alpha_sigma2_post, beta_sigma2_post = posterior_sigma2(y, X, mu_beta_post, alpha_star, beta_star)

# Print results
print("Posterior mean of beta:", mu_beta_post)
print("Posterior covariance of beta:\n", Sigma_beta_post)
print("Posterior parameters for sigma^2: alpha =", alpha_sigma2_post, ", beta =", beta_sigma2_post)
```

## Task Overview

- Model $\beta$ using `pm.Normal('betas', mu=0, sigma=1, shape=p)`.

- Perform Bayesian inference with MCMC diagnostics.

```python
import numpy as np
import pandas as pd
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Load or generate dataset
np.random.seed(42)
n, p = 100, 3  # 100 samples, 3 features
X = np.random.randn(n, p)
true_betas = np.array([2.5, -1.3, 0.7])
y = X @ true_betas + np.random.randn(n) * 0.5  # Add Gaussian noise

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Bayesian Linear Regression Model
with pm.Model() as model:
    # Priors
    betas = pm.Normal("betas", mu=0, sigma=1, shape=p)
    sigma = pm.HalfCauchy("sigma", beta=1)  # Non-inverse-gamma prior
    
    # Likelihood
    mu = pm.math.dot(X_train, betas)
    y_obs = pm.Normal("y_obs", mu=mu, sigma=sigma, observed=y_train)
    
    # Inference
    trace = pm.sample(2000, tune=1000, return_inferencedata=True)

# MCMC diagnostics
az.plot_trace(trace)
plt.show()

# Posterior summary
print(az.summary(trace))
```

```python
import numpy as np
import pandas as pd
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt
from scipy.stats import t as student_t
from sklearn.preprocessing import StandardScaler

# Generate synthetic data with outliers
np.random.seed(42)
n, p = 100, 3
X = np.random.randn(n, p)
true_betas = np.array([2.5, -1.3, 0.7])
y = X @ true_betas + np.random.randn(n) * 0.5

# Introduce outliers in response variable
outlier_indices = np.random.choice(n, size=5, replace=False)
y[outlier_indices] += np.random.randn(5) * 10  # Large deviations

# Standardize predictors
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Bayesian Robust Regression Model
with pm.Model() as robust_model:
    # Priors
    betas = pm.Normal("betas", mu=0, sigma=1, shape=p)
    nu = pm.Gamma("nu", alpha=2, beta=0.1)  # Prior for degrees of freedom in t-distribution
    tau = pm.Gamma("tau", alpha=nu / 2, beta=nu / 2, shape=n)  # Individual precisions
    
    sigma = pm.HalfCauchy("sigma", beta=1)  # Overall scale
    
    # Likelihood with Student-t heavy-tailed errors
    y_obs = pm.Normal("y_obs", mu=pm.math.dot(X_scaled, betas), sigma=sigma / pm.math.sqrt(tau), observed=y)
    
    # Inference
    trace = pm.sample(2000, tune=1000, return_inferencedata=True, target_accept=0.9)

# Posterior analysis
az.plot_trace(trace)
plt.show()

# Identify outliers using  (small  means high likelihood of being an outlier)
tau_post = trace.posterior["tau"].mean(dim=("chain", "draw"))
outlier_candidates = np.argsort(tau_post)[:5]  # Smallest  values

print("Potential outliers (smallest ):", outlier_candidates.values)
```
